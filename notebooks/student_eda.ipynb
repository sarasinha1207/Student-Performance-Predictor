# Student Performance Predictor - Exploratory Data Analysis (EDA)

This notebook explores the student performance dataset, checking for distribution of features, correlations, and relationship with the target variable (`final_score`).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
```

## 1. Load Dataset

```python
# Load synthetic dataset
df = pd.read_csv('../data/student_data.csv')
df.head()
```

## 2. Basic Dataset Info

```python
df.info()
```

```python
df.describe()
```

## 3. Exploratory Data Visualizations

### Target Variable Distribution: `final_score`

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['final_score'], kde=True, bins=20, color='royalblue')
plt.title('Distribution of Final Exam Scores')
plt.xlabel('Final Score')
plt.ylabel('Count')
plt.show()
```

### Numerical Correlations

```python
plt.figure(figsize=(10, 8))
# Select only numeric columns for correlation
numeric_cols = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Student Features')
plt.show()
```

### Study Hours vs Final Score

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='study_hours', y='final_score', data=df, alpha=0.7, color='teal')
plt.title('Study Hours vs Final Exam Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Final Score')
plt.show()
```

### Attendance vs Final Score

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='attendance', y='final_score', data=df, alpha=0.7, color='coral')
plt.title('Attendance vs Final Exam Score')
plt.xlabel('Attendance Percentage')
plt.ylabel('Final Score')
plt.show()
```

### Demographic Features Analysis: Parent Education Level vs Final Score

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='parent_education', y='final_score', data=df, palette='Set2')
plt.title('Final Exam Scores across Parent Education Levels')
plt.xlabel('Parent Education Level')
plt.ylabel('Final Score')
plt.xticks(rotation=15)
plt.show()
```
